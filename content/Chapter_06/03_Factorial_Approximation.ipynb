# Factorial Approximations

In [Sections 6.1](#ch6.1) and [6.2](#ch6.2) we introduced two methods for approximating exponentials:

$$e^x = \begin{cases} \sum_{n = 0}^{\infty} \frac{1}{n!} x^n & \approx 1 + x + \frac{1}{2} x^2 + \frac{1}{6} x^3 + ... \\ \lim_{n \rightarrow \infty} \left(1 + \frac{1}{n} x \right)^n & \approx \left(1 + \frac{1}{n} x \right)^n \text{ for large } n \end{cases} $$

Many probability calculations involve factorials. As discussed in [Section 6.2](#ch6.2), factorials are often hard to work with. This section introduces a method for approximating factorials using simpler functions. These approximations make it easier to answer questions like, *how big is $100!$?* *How big is $100$ choose $50$?*

Consider the definition of the factorial:

$$n!  = n \times (n - 1) \times (n - 2 ) \times ... 2 \times 1.$$

This product involves $n$ terms. The early terms all are of size $n$. So, to a very rough approximation, $n! \approx n^n$. Clearly $n! < n^n$ since all of the terms in the product except the first are less than $n$. Half are larger than $n/2$ and half are smaller, so a better approximation is $(n/2)^n$. Stirling's approximation provides an accurate estimate for $n!$ using a function roughly of the form $(n/e)^n$.

:::{note} Stirling's Approximation

$$n! \approx \sqrt{2 \pi e} \left(\frac{n}{e}\right)^{n + \frac{1}{2}}$$

The relative error in the approximation decays at order $\mathcal{O}(n^{-1})$. 

:::

:::{note} Factorials Grow Superexponentially!

Stirling's formula shows that $n!$ grows *faster than* exponential. Applying the limiting method:

$$\begin{aligned} \lim_{n \rightarrow \infty} \frac{a^n}{n!} & = \lim_{n \rightarrow \infty}  \frac{1}{\sqrt{2 \pi e}}\frac{a^n}{(n/e)^{n + 1/2}} \\ &  =  \frac{1}{\sqrt{2 \pi}} \lim_{n \rightarrow \infty}  \frac{1}{n^{1/2}}\left( \frac{a e}{n} \right)^n. \end{aligned}$$

This limit converges to zero for any $a$ since $(a \times e)/n$ converges to zero as $n$ grows. It follows that $\lim_{n \rightarrow \infty} \frac{a^n}{n!} = 0$ for all $a$, so $n!$ diverges *faster* than $a^n$ for any base $a$. 

:::

## Application: Sampling Without Replacement

Suppose you sample $m$ items uniformly, without replacement, from a fixed pool of $n$ items. Suppose that the $n$ items can be categorized into two types, and that $k$ of the items are of type 1. Let $X$ denote the number of items in your sample of type 1. What is the chance that $X = x$?

For example, you draw $m = 13$ cards from a standard deck of $n = 52$, of which $k = 16$ are face cards and $n - k = 36$ are not face cards. What is the chance that your draw includes 7 face cards?

The chance of selecting $x$ items of type 1 from the pool is found using probability by proportion (for a derivation, see [Section 4.2](#ch4.2)):

$$\text{Pr}(X = x) =  \frac{\left( \begin{array}{c} k \\ x \end{array} \right) \left( \begin{array}{c} n - k \\ m - x \end{array} \right)}{\left( \begin{array}{c} n \\ m \end{array} \right)}.$$
