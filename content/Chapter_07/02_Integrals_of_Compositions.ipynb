# Integrals of Compositions

In [Section 3.2](#ch3.2) we introduced the idea of a function composition. 

:::{note} Function Composition

A **composition** of two functions, $g \circ h$ is the function:

$$f(x) = g(h(x)). $$

:::

We've seen a variety of distributions that involve a composition, most importantly, the normal distribution, whose PDF has the form:

$$\text{PDF}(x) \propto e^{-\frac{1}{2} x^2}.$$

The simplest function compositions use a linear function for $h$:

$$f(x) = g(h(x)) = g((x -  s)/\sigma), \quad h(x) = (x - s)/\sigma $$

translates the function by $s$, and dilates it by $\sigma$. Most distribution families are defined by first picking some family of functions $g$, then allowing arbitrary linear $h$. Then the parameters $s$ and $\sigma$ of $h$ act as location and scale parameters that can be adjusted to translate and dilate the base model, $g$. 

## Integration by Substitution (Change of Variables)

Consider an integral of a function composition:

$$\int_{x = a}^b g(h(x)) dx. $$

In [Section 7.1](#ch7.1) we worked out rules for integrating sums and products of functions. What about nested functions?

The procedure used to integrate a function composition is often called "$u$" substitution after the convention $u = h(x)$. There is no reason to call the output of the inner function $u$. So, we won't call this rule "$u$" substitution. Instead, we'll name it based on the concept that justifies the rule: *change of variables*

:::{note} Integration by *Change of Variables* (Version I)

Suppose that $f(x) = g \circ h(x) = g(h(x))$ and that $h(x)$ is a monotonic function.

Let $u = h(x)$ and $du(x) = h'(x) dx$. Then:

$$ \int g(h(x)) h'(x) dx = \int g(u) du. $$

Let $G$ denote the indefinite integral of $g$. Then:

$$ \int_{a}^b g(h(x)) h'(x) dx  = G(u) \Big|_{h(a)}^{h(b)} = G(h(a)) - G(h(b)).$$

This procedure is equivalent to *changing variables by replacing $x$ with $u = h(x)$*. 

:::{caution} Bounds of Integration

Make sure you always check the bounds of integration. You need to update them by replacing the bounds on $x$ with the bounds on $u = h(x)$. 

:::



:::{hint} Proof

The proof follows from the fundamental theorem of calculus and the chain rule for derivatives. 

First, let $F(x) = G \circ h(x) = G(h(x))$ where $G$ is the antiderivative of $g$, $\frac{d}{dx} G(x) = G'(x) = g(x)$. Then, by the product rule:

$$F'(x) = \frac{d}{dx} F(x) = G'(h(x)) h'(x) = g(h(x)) h'(x).$$

So, by the Fundamental Theorem of Calculus:

$$F(x) = \int F'(x) dx = \int g(h(x)) h'(x) dx. $$

Then:

$$G(h(a)) - G(h(b)) = F(a) - F(b) = \int_{x = a}^b F'(x) dx = \int_{x = a}^b g(h(x)) h'(x) dx. $$


:::

:::{tip} Example

Suppose that $X \sim \text{Normal}(0,1)$. Then, since $X$ is a standard normal random variable, $\mathbb{E}[X] = 0$ and $\text{SD}[X] = 1$. Find the mean absolute deviation in $X$.

First, write down the expectation as a weighted average:

$$\text{MAD}[X] = \mathbb{E}[|X - 0|] = \int_{x = -\infty}^{\infty} |x| \text{PDF}(x) dx = \frac{1}{\sqrt{2 \pi}} \int_{x = -\infty}^{\infty} |x| e^{-\frac{1}{2} x^2} dx. $$

Next, let's think a bit about symmetries. The normal density $\text{PDF}(x) \propto e^{-\frac{1}{2} x^2}$ is an even function. The function $|x|$ is also even since $|-x| = |x|$. The product of two even functions is an even function. Therefore:

$$\text{MAD}[X] = \frac{2}{\sqrt{2 \pi}} \int_{x = 0}^{\infty} x e^{-\frac{1}{2} x^2} dx.$$

Notice that $e^{-\frac{1}{2} x^2}$ is a composition of the functions $g(x) = e^{-x}$ and the quadratic function $h(x) = \frac{1}{2} x^2$. 

So, let's try integrating by changing variables.

Let $u = h(x) = \frac{1}{2} x^2$. Then $du = h'(x) dx = \frac{2}{2} x dx = x dx$. So:

$$\begin{aligned} \text{MAD}[X] & = \frac{2}{\sqrt{2 \pi}} \int_{x = 0}^{\infty} x e^{-\frac{1}{2} x^2} dx \\
& = \frac{2}{\sqrt{2 \pi}} \int_{x = 0}^{\infty} e^{-\frac{1}{2} x^2} x dx \\
& = \frac{2}{\sqrt{2 \pi}} \int_{x = 0}^{\infty} g(h(x))  h'(x) dx \\
& = \frac{2}{\sqrt{2 \pi}} \int_{u = h(0)}^{h(\infty)} g(u) du \\
& =  \frac{2}{\sqrt{2 \pi}} \int_{u = 0}^{\infty} e^{-u} du. \end{aligned}$$

Pay careful attention to the bounds of integration. In the end, we still integrate from 0 to infinity since $h(0) = \frac{1}{2} 0^2 = 0$ and $\lim_{x \rightarrow \infty} h(x) = \frac{1}{2} \lim_{x \rightarrow \infty} x^2 = \infty$. 

Integrating:

$$ \int_{u = 0}^{\infty} e^{-u} du = -e^{-u} \Big|_{0}^{\infty} = -(0 - 1) = 1.$$

Therefore:

$$\text{MAD}[X] = \frac{2}{\sqrt{2 \pi}} = \sqrt{\frac{2}{\pi}} \approx \sqrt{\frac{2}{3}} \approx 0.80. $$

Notice that, as Jensen guarantees, $\text{MAD}[X] \approx \sqrt{2/3} < \sqrt{1} = 1 = \text{SD}[X]$. 

:::


Often, we want to just take an integral of a composition, e.g. $\int_{x = a}^b g(h(x)) dx$. In this case we are missing the $h'(x) dx$ term needed to replace $g(h(x)) dx$ with $g(u) du$. So, we add it in ourselves:

$$g(h(x)) dx = g(h(x)) \frac{h'(x)}{h'(x)} dx = \frac{g(h(x))}{h'(x)} h'(x) dx. $$

Now, to write everything in terms of $u$, we can substitute $h(x) = u$, $h'(x) dx = du$, but get stuck with $h'(x)$. If $h$ is monotonic, then $h$ is invertible over its range, so we can define an *inverse function* $h^{-1}$ such that $h^{-1}(h(x)) = x$ and $h(h^{-1}(u)) = u$. Then, given $u = h(x)$, $x = h^{-1}(u)$. So, we can replace $h'(x)$ with $h'(h^{-1}(u))$. 

Then:

$$g(h(x)) dx = \frac{g(h(x))}{h'(x)} h'(x) dx = \frac{g(u)}{h'(h^{-1}(u))} du. $$

Integrating returns the standard form for integration by change of variables:

$$\int g(h(x)) dx = \int \frac{g(u)}{h'(h^{-1}(u))} du. $$

Then:

$$\int_{x = a}^b g(h(x)) dx = \int_{u = h(a)}^{h(b)} \frac{g(u)}{h'(h^{-1}(u))} du. $$

Let $\tilde{g}(u) = g(u)/h'(h^{-1}(u))$. Then let $\tilde{G}$ represent the anti-derivative (indefinite integral) of $\tilde{g}$. Then:

$$\int_{x = a}^b g(h(x)) dx = \int_{u = h(a)}^{h(b)} \tilde{g} du = \tilde{G}(h(a)) - \tilde{G}(h(b)). $$

:::{note} Integration by *Change of Variables* (Version II)

Suppose that $f(x) = g \circ h(x) = g(h(x))$ and that $h(x)$ is a monotonic function.

Let $u = h(x)$ and $du(x) = h'(x) dx$. Let $h^{-1}$ denote the inverse function that recovers $x$ from $u$, $x = h^{-1}(u)$. 

Then:

$$ \int g(h(x)) dx = \int \frac{g(u)}{h'(h^{-1}(u))} du = \int \tilde{g}(u) du. $$

Let $\tilde{G}$ denote the indefinite integral of $\tilde{g}(x) = g(u)/h'(h^{-1}(u))$. Then:

$$ \int_{a}^b g(h(x)) dx  = \tilde{G}(u) \Big|_{h(a)}^{h(b)} = \tilde{G}(h(a)) - \tilde{G}(h(b)).$$

:::

This form is a bit messy as a single formula. It's easier to remember as a procedure:

1. Identify a term in the integrand we would like to replace (e.g. a function of $x$, $h(x)$, that we will use to define a new variable, $u$).

1. Replace $h(x)$ with $u$ everywhere inside the integral.

1. Find $h'(x) = \frac{d}{dx} h(x)$. 

1. Replace $dx$ with $du/h'(x)$ inside the integral. 

1. Solve for $x$ in terms of $u$, $x = h^{-1}(u)$. 

1. Replace all remaining $x$'s inside the integral with $h^{-1}(u)$. 

1. Update the bounds of integration from $x \in [a,b]$ to $u \in [h(a),h(b)]$.

1. Integrate.

:::{tip} Example

Suppose that $X \sim \text{Normal}(0,1)$. Find $\mathbb{E}[|X|^3]$. 

As before, start by writing the expectation as a weighted average:

$$\mathbb{E}[|X|^3] = \frac{1}{\sqrt{2 \pi}} \int_{x = -\infty}^{\infty} |x|^3 e^{-\frac{1}{2} x^2} dx. $$

Then, since $|x|^3$ is the composition of an odd function, $x^3$, with an even function, $|x|$, the function $|x|^3$ is even. The normal density is also even, so $|x|^3 e^{-\frac{1}{2} x^2}$ is an even function. So, using the same symmetry argument we used before:

$$\mathbb{E}[|X|^3] = \frac{2}{\sqrt{2 \pi}} \int_{x = 0}^{\infty} x^3 e^{-\frac{1}{2} x^2} dx. $$

As before, we'd like to set $u = h(x) = \frac{1}{2} x^2$ and $g(x) = e^{-x}$. However, this time the term outside $g(h(x))$ is $x^3 \neq h'(x) = \frac{d}{dx} \frac{1}{2} x^2 = x$. So we need to use the second version of integration by change of variables.

1. $u = h(x) = \frac{1}{2} x^2$.

1. Update the integral:

    $$\frac{2}{\sqrt{2 \pi}}  \int_{x = 0}^{\infty} x^3 e^{-u} dx $$

1. $h'(x) = \frac{d}{dx} \frac{1}{2} x^2 = x$. So, $du = x dx$.

1. Replace $dx$ with $du/h'(x)$ inside the integral:

    $$\frac{2}{\sqrt{2 \pi}}  \int_{x = 0}^{\infty} x^3 e^{-u} \frac{x dx}{x} = \frac{2}{\sqrt{2 \pi}}  \int_{x = 0}^{\infty} x^3 e^{-u} \frac{du}{x} $$

    Simplifying:

    $$\frac{2}{\sqrt{2 \pi}}  \int_{x = 0}^{\infty} x^2 e^{u} du $$

1. Solve for $x$ in terms of $u$. 

    $$u = h(x) = \frac{1}{2} x^2 \Rightarrow x = \pm \sqrt{2 u} $$

    Since we are only integrating over positive $x$, we can assume that $x = \sqrt{2 u}$ so $h^{-1}(u) = \sqrt{2 u}$.


1. Replace all remaining $x$'s inside the integral with $h^{-1}(u)$. 

    $$\frac{2}{\sqrt{2 \pi}}  \int_{x = 0}^{\infty} (\sqrt{2 u})^2 e^{u} du = \frac{2}{\sqrt{2 \pi}} \int_{x = 0}^{\infty} 2 u e^{u} du $$

    Simplifying again:

    $$\frac{4}{\sqrt{2 \pi}} \int_{x = 0}^{\infty} u e^{-u} du. $$

1. Update the bounds of integration from $x \in [a,b]$ to $u \in [h(a),h(b)] = [\frac{1}{2} 0^2, \frac{1}{2} \infty^2] = [0,\infty]$.

    $$\frac{4}{\sqrt{2 \pi}} \int_{u = 0}^{\infty} u e^{-u} du. $$

1. Integrate. Notice that the integrand is a product, $u \times e^{-u}$. So, we'll use integration by parts (see [Section 7.1](#ch7.1)):

    $$\begin{aligned} \int_{u = 0}^{\infty} u e^{-u} du & = -ue^{-u} |_{0}^{\infty} - \int_{u = 0}^{\infty}(-e^{-u}) du \\
    & = (0 - 0) + \int_{u=0}^{\infty} e^{-u} du \\
    & = -e^{-u} |_{0}^1 = -(0-1) = 1. \end{aligned}$$

Therefore:

$$\mathbb{E}[|X|^3] = \frac{4}{\sqrt{2 \pi}} = 2 \sqrt{\frac{2}{\pi}}. $$

:::

## Change of Density

The rule we just worked out for changing variables inside of an integral provides a general rule for updating the density of a continuous random variable after transforming the random variable. 

Suppose that $X$ is a continuous random variable, with density function $f_X(x)$. Then to find the probability that $X$ lands in an interval, we would evaluate an integral over the density:

$$\text{Pr}(X \in [a,b]) = \int_{x = a}^b f_X(x) dx.$$

Similarly, to find the CDF we would integrate:

$$F_X(x) = \text{Pr}(X \in (-\infty,x]) = \int_{s = -\infty}^x f_X(s) ds. $$

Suppose now, that $Y = h(X)$ for some monotonically increasing, differentiable function $h$. What is the density of $Y$, $f_Y(y)$?

Well, just like $X$, the chance that $Y$ lands in an interval is related to its density by an integral:

$$\text{Pr}(Y \in [c,d]) = \int_{x = c}^d f_Y(y) dy.$$

The CDF of $Y$ is also related to its density by an integral:

$$F_Y(y) = \text{Pr}(Y \in (-\infty,y]) = \int_{s = -\infty}^y f_Y(s) ds. $$

There are now two ways to find the density of $Y$:

1. Start from integrals involving $x$, and use integration by change of variables to replace $x$ with $y = h(x)$. 

1. Start from integrals involving $y$, and try to re-express them in terms of $x$. Then, match sides to solve for the density of $y$, and check that integrating over $y$ gives the same answer as integrating over $x$.

We will adopt the second approach.

Consider the CDF of $Y$. As always, if we know the CDF, then we can recover the density, and chances on intervals. So, if we can work out the CDF of $Y$, then we have recovered it's distribution. In particular, if we know the CDF of $Y$, then we can find its density since:

$$f_Y(y) = \frac{d}{dy} F_Y(y). $$

The CDF of $Y$ is:

$$F_Y(y) = \text{Pr}(Y \leq y) = \text{Pr}(h(X) \leq y). $$

Recall that, if $h(x)$ is monotonically increasing, then $h(x) \leq y$ if and only if $x \leq h^{-1}(y)$. Therefore:

$$F_Y(y) = \text{Pr}(X \leq h^{-1}(y)) = F_X(h^{-1}(y)). $$

So, recalling the product rule, and that $\frac{d}{dx} F_X(x) = f_X(x)$,

$$f_Y(y) = \frac{d}{dy} F_X(h^{-1}(y)) = f_X(h^{-1}(y)) \frac{d}{dy} h^{-1}(y). $$

So:

:::{note} Change of Density Formula

If $X$ is a continuous random variable with density $f_X$, and $Y = h(X)$ for some differentiable, monotonically increasing function $h$, then:

$$f_Y(y) = f_X(h^{-1}(y)) \frac{d}{dy} h^{-1}(y).$$

It is often helpful to write this result in terms of $x$. Let $h'(x) = \frac{d}{dx} h(x)$. Then, 

$$\frac{d}{dy} h^{-1}(y) = \frac{1}{h'(x)} \text{ at } x = h^{-1}(y) $$

So:

$$f_Y(y) = f_X(x) \frac{1}{h'(x)} \text{ at }  x = h^{-1}(y). $$

If $h$ is differentiable, but not monotonic, then:

$$f_Y(y) = f_X(x) \frac{1}{|h'(x)|} \text{ at }  x = h^{-1}(y). $$

This formula is the **change of density formula.**

The absolute value enters since densities are nonnegative and the $1/h'(x) dx$ term represents the length of an interval. Lengths can never be negative. We didn't need the absolute value in the increasing case since, if $h$ is increasing, then $h'(x) > 0$ so $|h'(x)| = h'(x)$. 


:::{hint} Derivatives of Inverses
:class: dropdown

Why is $\frac{d}{dy} h^{-1}(y)$ equal to $1/h'(x)$ at $x = h^{-1}(y)$?

Well:

$$h(h^{-1}(y)) = y $$

so:

$$\frac{d}{dy} h(h^{-1}(y)) = \frac{d}{dy} y = 1. $$

Now, applying the product rule:

$$\frac{d}{dy} h(h^{-1}(y)) = h'(h^{-1}(y)) \left[ \frac{d}{dy} h^{-1}(y) \right]. $$

So, matching terms:

$$ h'(h^{-1}(y)) \left[ \frac{d}{dy} h^{-1}(y) \right] = 1$$

or:

$$ \frac{d}{dy} h^{-1}(y) = \frac{1}{h'(h^{-1}(y))}. $$

Geometrically, $h^{-1}$ is the reflection of $h$ across the line $y = x$. So, since the slope of $h$ is change in $y = h(x)$ over change in $x$, the slope of $h^{-1}$ is the change in $x$ over the change in $y$, which is one over the slope of $h$. 

:::

:::{hint} 

The change of density formula can be intimidating, and hard to remember. To use it, treat it as a procedure. Always:

1. Write down the transformation $Y = h(X)$

1. Solve for $x$ in terms of $y$ to get $x = h^{-1}(y)$.

1. Solve for $h'(x) = \frac{d}{dx} h(x)$.

1. Substitute in. 

If you forget how to change densities, set up the integral for the CDF instead, then apply integration by substitution to integrate over $y$ instead of $x$. Proceed using the "u-substitution" rule until all $x$ terms in the integral have been replaced with $y$'s. The function inside the integral is now the density of $Y$. 

:::


We've already used a special case for a while. 

Suppose that $h(x)$ is linear, $h(x) = \sigma x + s$ for some $\sigma > 0$. Then $h'(x) = sigma$, so, by the change of density formula:

$$f_Y(y) \propto f_X(h^{-1}(y)) = f_X\left( \frac{x - s}{\sigma} \right). $$

So $Y = h(X)$ has density proportional to the density of $X$ translated by $s$ and dilated by $\sigma$. To maintain normalization, we've noticed that, if we dilate a distribution by a factor of $\sigma$, then we must divide its height by $\sigma$ so that it integrates to one. As before, if we double the width of a rectangle, we have to half its height to keep its area fixed.

By that logic, we should have:

$$f_Y(y) = \frac{1}{\sigma} f_X\left( \frac{x - s}{\sigma} \right)$$

Using the exact change of density formula gives the same result since $h'(x) = \sigma$ for all $x$:


$$f_Y(y) = f_X(h^{-1}(y)) \frac{1}{|h'(h^{-1}(y))|} = f_X\left( \frac{x - s}{\sigma} \right) \frac{1}{|\sigma|} . $$


:::{note} Linear Change of Density

If $Y = a X + b$, and $X$ is continuously distributed, then:

$$f_Y(y) = \frac{1}{|a|} f_X\left(\frac{y - b}{a} \right). $$

:::

This is a sensible rule. Replacing $X$ with $2X$ doubles the distance between any sampled values of $X$, so should halve its density.

:::{tip} Example

Suppose that $X \sim \text{Normal}(0,1)$ and $Y = \sigma X + \mu$. Then:

$$f_Y(y) = \frac{1}{\sigma} f_X((x - \mu)/\sigma) = \frac{1}{\sigma} \frac{1}{\sqrt{2 \pi}} e^{-\frac{1}{2} \left(\frac{x - \mu}{\sigma} \right)^2} $$

The function on the right is the density function for a **generic normal random variable** or a **Gaussian** random variable, $Y \sim \text{Normal}(\mu,\sigma)$. 
:::

Let's check that the change of density formula actually returns the correct density for $Y$. To confirm, we'll use integration by substitution. We will check the case when $h$ is monotonically increasing. To check the general case, break the range of $x$ into segments where $h$ is monotonic, and work one segment at a time.

As long as $h$ is increasing:

$$\text{Pr}(X \in [a,b]) = \text{Pr}(Y \in [h(a),h(b)]).$$

So, integrating over the density of $Y$:

$$\text{Pr}(Y \in [h(a),h(b)]) = \int_{y = h(a)}^h(b) f_Y(y) dy.$$

Using the change of density formula:

$$\int_{y = h(a)}^h(b) f_Y(y) dy = \int_{y = h(a)}^h(b) f_X(h^{-1}(y)) \left[\frac{d}{dy} h^{-1}(y) \right] dy.$$

That looks like a mess, but we have all the parts we need to use integration by change of variables. 

Let $x = h^{-1}(y)$. Then:

1. $f_X(h^{-1}(y)) = f_X(x)$,

1. $dx = \frac{d}{dy} h^{-1}(y) dy $$

1. $h^{-1}(h(a)) = a$ and $h^{-1}(h(b)) = b$. 

So:

$$\begin{aligned} \text{Pr}(X \in [a,b]) & = \int_{y = h(a)}^h(b) f_X(h^{-1}(y)) \left[\frac{d}{dy} h^{-1}(y) \right] dy \\
& = \int_{x = a}^b f_X(x) dx = \text{Pr}(X \in [a,b]). \end{aligned} $$

:::{tip} Example
:class: dropdown

Suppose that $X \sim \text{Exp}(\lambda)$ and $Y  = \sqrt{x}$. 

Then, let $h(x) = x^{1/2}$. It follows that:

1. $h'(x) = \frac{1}{2} x^{-1/2}$
1. $h^{-1}(y) = y^2$

So:

$$f_Y(y) = \frac{1}{|h'(x)|} f_X(x) = 2 x^{1/2} \lambda e^{-\lambda x}  \text{ at } x = h^{-1}(y) = y^2.$$

Plugging in:

$$f_Y(y) = 2 \lambda y^2 e^{-\lambda y^2}.$$

The support of $Y$ is $[0,\infty)$ since the smallest possible value of $\sqrt{X}$ is zero, and the largest is infinity.

:::

:::{tip} Example
:class: dropdown

Suppose that $X \sim \text{Exponential}(0,1)$ and $Y  = 1/x$. 

Then, let $h(x) = x^{1/2}$. It follows that:

1. $|h'(x)| = |-x^{-2}| = x^{-2}$
1. $h^{-1}(y) = 1/y = y^{-1}$.

So:

$$f_Y(y) = \frac{1}{|h'(x)|} f_X(x) = \frac{1}{x^2} \lambda e^{-\lambda x}  \text{ at } x = h^{-1}(y) = \frac{1}{y}.$$

Plugging in:

$$f_Y(y) = \lambda y^2 e^{-\frac{\lambda}{y}}.$$

The support of $Y$ is $[0,\infty)$ since the smallest possible value of $1/X$ is zero ($X \rightarrow \infty$), and the largest is infinity ($X \rightarrow 0$).

:::

:::{tip} Example
:class: dropdown

Suppose that $X \sim \text{Uniform}(0,1)$ and $Y  = -\log(X)$. 

Then, let $h(x) = -\log(x)$. It follows that:

1. $|h'(x)| = |-x^{-1}| = x^{-1}$
1. $h^{-1}(y) = e^y$.

So:

$$f_Y(y) = \frac{1}{|h'(x)|} f_X(x) = \frac{1}{x} \times 1  \text{ at } x = h^{-1}(y) = e^y.$$

Plugging in:

$$f_Y(y) = e^{-y}.$$

The support of $Y$ is $[0,\infty)$ since the smallest possible value of $-\log(X)$ is zero ($X = 1$), and the largest is infinity ($X \rightarrow 0$).

So, $Y \sim \text{Exponential}(1)$. 

:::

:::{tip} Example
:class: dropdown

Suppose that $X \sim \text{Normal}(0,1)$ and $Y  = X^2$. 

Then, let $h(x) = x^2$. Now we have a problem since $h$ is not monotonic. However, since the density of $X$ is an even function, and $h$ is an even function, the density of $Y$ would not change had we conditioned on $X > 0$. On only positive $x$, $x^2$ is monotonic. So, we can use our familiar rule, replacing $f_{X}(x)$ with $f_{X|X > 0}(x)$.

Then:

1. $|h'(x)| = |2x| = 2x$
1. $h^{-1}(y) = \sqrt{y}$.

So:

$$f_Y(y) = \frac{1}{|h'(x)|} f_{X|X > 0}(x) = \frac{1}{2 x} \times \frac{2}{\sqrt{2 \pi}} e^{-\frac{1}{2} x^2}  \text{ at } x = h^{-1}(y) = \sqrt{y}.$$

Plugging in:

$$f_Y(y) = \frac{1}{\sqrt{2 \pi}} y^{-1/2} e^{- \frac{1}{2} y}.$$

The support of $Y$ is $[0,\infty)$ since the smallest possible value of $X^2$ is zero ($X = 0$), and the largest is infinity ($X \rightarrow \infty$).

:::
