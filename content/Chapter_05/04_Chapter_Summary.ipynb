# Chapter Summary

## Tails

Definitions and examples are all available in [Section 5.1](#ch5.1).

1. The **tail** of a distribution is the part of a distribution associated with unusually extreme events. 

1. The **survival function** is the function $\text{Pr}(X > x)$ as a function of $x$. 

1. We derived the survival function and CDF of a geometric random variable.

1. We derived the value of a **geometric series**: $\sum_{n = 0}^{\infty} r^n = (1 - r)^{-1}$ if $|r| < 1$.

1. We showed that the **harmonic series**: $\sum_{n = 1}^{\infty}n^{-1}$ diverges. 

## Asymptotic Rates

Definitions and examples are all available in [Section 5.2](#ch5.2).

1. A **sequence** is an infinitely long, order list, of numbers.

1. We can visualize convegergence rates using **log-log plots**, which convert power laws, $a(n) = c n^{-\gamma}$ into lines whose slope is the exponent, $-\gamma$.

1. We can compare how quickly series converge to zero, or diverge to infinity, by studying the limiting ratio:

    $$L = \lim_{n \rightarrow \infty} \frac{a(n)}{b(n)} $$

    - If $L = 0 $ then $a(n)$ converges to zero faster than $b(n)$
    - If $0 < L < \infty$ then $a(n)$ converges at the same rate as $b(n)$
    - If $L = \infty$ then $a(n)$ slower than $b(n)$

    - If $a(n)$ is faster than, or as fast as, $b(n)$ then $a(n) = \mathcal{O}(b(n))$ (is *on the order of* $b(n)$)

1. Sometimes it is helpful to evaluate the limit using **L'Hopital's Rule** (e.g. use the limiting ratio of the derivatives)

1. Exponential sequences, $r^n$ for $|r| < 1$ converge *faster than* all power laws $n^{-\gamma}$ for $\gamma > 0$.

1. A distribution whose tails decay slowly (e.g. power laws) is **heavy tailed.**


## Convergence Tests for Series

Definitions and examples are all available in [Section 5.3](#ch5.3).

We considered three tests:

1. **The Integral Comparison Test**

    - Compare the series to a related integral. The series converges if and only if the integral converges.

1. **The Direct and/or Limit Comparison Test**

    - Compare the sequences $a$ and $b$. 
    - If $a$ converges to zero faster than, or as fast as, $b$, and the series defined by $b$ converges, then so does the series defined by $a$.
    - If $a$ converges to zero slower than, or as slowly as, $b$, and the series defined by $b$ diverges, then so does the series defined by $a$.

1. **The Ratio Test**

    - Evaluate the limiting ratio: $r = \lim_{n \rightarrow \infty} \frac{|a(n+1)|}{|a(n)|}$ to compare to a geometric series.
