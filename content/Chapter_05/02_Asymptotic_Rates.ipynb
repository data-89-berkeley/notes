# Asymptotic Rates

In [Section 5.1](#ch5.1) we saw that two sequences converging to zero can appear quite similar, but behave differently. Both the geometric sequence, $\{r^n\}_{n=0}^{\infty}$ for $0 < r < 1$, and the harmonic sequence, $\{\frac{1}{n}\}_{n=1}^{\infty}$, decay monotonically and converge to zero as $n$ diverges. However, the geometric *series*, $\sum_{n=0}^{\infty} r^n = 1/(1 - r)$ converges while the harmonic *series*, $\sum_{n=1}^{\infty} \frac{1}{n}$ diverges. These two sequences are distinguished by the *rate* at which they approach zero. 

This section introduces the visual, and analytic, tools needed to compare the rates at which sequences approach zero in a limit. Since these rates apply in a limit, they are often called *asymptotic* rates, where *asymptotic* stands for, "regarding approach to an asymptote, e.g. a limit." We will use asymptotic rates to categorize the tail behavior of distributions. Distributions whose tails decay quickly are "light" tailed, while distributions whose tails decay slowly are "heavy" tailed. The heavier the tail, the more of the mass of the distribution is spread out over unusually extreme outcomes. 

First, some definitions to standardize our language.

:::{note} Sequences Definition

A **sequence** is an infinitely long, ordered list of numbers.

:::

For example, $\{1,2,3,4,5,...\}$ is the sequence of all integers, while $\{1,4,9,16,25,...\}$ is the sequence of all squared numbers. 

Sequences are commonly expressed:

$$\{a(n)\}_{n=1}^{\infty} = \{a(1),a(2),a(3),...\} $$

where $a(n)$ is a function that returns the $n^{th}$ entry in the sequence. For example, the sequence consisting of all squared integers could be expressed $\{n^2\}_{n=1}^{\infty}$.

Discrete distribution functions for count variables are naturally expressed using an ordered list. If $X \in \{1,2,3,4,...,\infty\}$ then we can express its PMF as a sequence:

$$\{\text{PMF}(x)\}_{x=1}^{\infty} = \{p(x)\}_{x=1}^{\infty} $$

where $p(x) = \text{Pr}(X = x)$. For example, if $X$ is geometric, then $p(x) = (1 - p)^{x - 1} p$. 

The **tail decay rate** of a discrete distribution is the rate at which the sequence of values $p(x)$ decays as $x$ diverges. 

## Visualizing Tail Decay

Consider an example geometric and harmonic series again:

$$a(n) = \left(\frac{6}{7} \right)^n, \quad b(n) = n^{-1} $$

The geometric series converges to zero faster, but it is hard to see if we plot each series as a function of $n$, since the geometric is slower at first, and, by the time it catches the harmonic, both are already very small. 

![Geometric vs Harmonic.](geometric_vs_harmonic_sequences.png "Geometric and Harmonic Sequences") 


The same problem shows up anytime we want to compare the tails of a distribution. For example, if $X$ has power law tails, then we could have $p(x) \propto x^{-\gamma}$ for any $\gamma > 1$. Does $x^{-2}$ converge faster or slower than $x^{-3}$? How does the tail decay rate depend on the power of the power law, $\gamma$?

Once again, a plot is not too helpful, since the action is all squeezed to close to zero for the eye to discern. 

For specificity, let's focus on three power law type sequences:

$$a(n) = n^{-1}, \quad b(n) = n^{-2}, \quad c(n) = n^{-3}. $$

It shouldn't be too susrprising that the larger the negative power, the faster the sequence decays. Here's a table comparing values for inputs spaced as powers of ten.

$n$| $10^0 = 1$  | $10^1 = 10$ | $10^2 = 100$  | $10^3 = 1000$  |
:----:|:-------------|:-------------| :-------------| :-------------|
$a(n) = n^{-1}$  | 1 | 0.1 | 0.01 | 0.001 |
$b(n) = n^{-2}$  | 1 | 0.01 | 0.001 | 0.00001 |
$c(n) = n^{-3}$  | 1 | 0.001 | 0.000001 | 0.000000001 |

Notice that the last sequence converges *much* faster than the first. The first sequence adds one zero before the "1" each time we increase $n$ by a factor of ten. The second adds two zeros per factor of ten. The third adds three zeros per factor of ten. Notice, the number of zeros added matches the power of the power law. 

Let's write the table again, this time using scientific notation:

$n$| $10^0$  | $10^1$ | $10^2$  | $10^3$  |
:----:|:-------------|:-------------| :-------------| :-------------|
$a(n) = n^{-1}$  | $10^{0}$  | $10^{-1}$ | $10^{-2}$ | $10^{-3}$ |
$b(n) = n^{-2}$  | $10^{0}$  | $10^{-2}$ | $10^{-4}$ | $10^{-6}$ |
$c(n) = n^{-3}$  | $10^{0}$  | $10^{-3}$ | $10^{-6}$ | $10^{-9}$ |

Finally, only report the exponents by changing the header from $n$, to $log_{10}(n)$ and from the sequence values to the $\log_{10}$ of the sequence values:

$\log_{10}(n)$| $0$  | $1$ | $2$  | $3$  |
:----:|:-------------|:-------------| :-------------| :-------------|
$\log_{10}(a(n)) =- 1 \times \log_{10}(n)$  | 0  | -1 | -2 | -3 |
$\log_{10}(a(n)) =- 2 \times \log_{10}(n)$   | 0  | -2 | -4 | -6 |
$\log_{10}(a(n)) =- 3 \times \log_{10}(n)$   | 0  | -3 | -6 | -9 |

Converting to a log scale on the input and output makes it much easier to compare rates of convergence. Here's a corresponding plot using a log scale on both the input and output axes.


![Power Law Comparison.](log_log.png "Power Law Comparison on a Log-Log Plot") 


Notice, on the log-log plot, the power laws have become lines, whose slopes match their powers. The larger the negative power, the steeper the line on the log-log plot. 

:::{tip} Power Laws and Log-Log Plots

Given $a(n) = c n^{-\gamma}$:

$$\log(a(n)) = \log(c) + \log(n^{-\gamma}) = -\gamma \times \log(n) + \log(c) $$

so $\log(a(n))$ is a linear function of $\log(n)$ with slope $-\gamma$ and intercept $\log(c)$. 

:::

Log-log plots are very helpful tools for distinguishing rates of convergence. They are a good choice for visualizing tail decay since they convert power laws to lines, whose slope equals the power of the power line.

## Comparing Asymptotic Rates

Formally, we can compare the rate at which two sequences do to zero as follows:

:::{note} Convergence Rates

Given two sequences, $\{a(n)\}_{n=1}^{\infty}$ and $\{b(n)\}_{n=1}^{\infty}$ that both converge to zero, compute:

$$L = \lim_{n \rightarrow \infty} \frac{a(n)}{b(n)} $$

1. If $L = 0$ then $a(n)$ converges *faster* than $b(n)$
1. If $0 < L < \infty$ then $a(n)$ converges *at the same rate* than $b(n)$
1. If $L =  \infty$ then $a(n)$ converges *slower* than $b(n)$

:::

For example, $n^{-2}$ converges faster than $n^{-1}$ since $n^{-2}/n^{-1} = n^{-1}$ and $n^{-1}$ converges to zero as $n$ diverges. 

Less trivially, consider the example sequences:

$$a(n) = \frac{n}{1 + n^2}, \quad b(n) = \frac{1}{n - 1}. $$

Taking their ratio:

$$\frac{a(n)}{b(n)} = \frac{n (n - 1)}{1 + n^2} = \frac{n^2 - n}{n^2 + 1} $$

Then, we want to find the value of the limit:

$$L = \lim_{n \rightarrow \infty} \frac{n^2 - n}{n^2 + 1}.$$

There are two ways to evaluate this limit. First, try simplifying. Divide the top and the bottom by $n^2$:

$$\begin{aligned} L & = \lim_{n \rightarrow \infty} \frac{(n^2 - n)/n^2}{(n^2 + 1)/n^2} = \lim_{n \rightarrow \infty} \frac{1 - n^{-1}}{1 + n^{-2}} \\
&  = \frac{1 - \lim_{n \rightarrow \infty} n^{-1}}{1 + \lim_{n \rightarrow \infty} n^{-2}} = \frac{1 - 0}{1 + 0} = 1. \end{aligned}$$

Alternately, recognize that $n^2$ diverges *faster* than $n$, so for large $n$ the $n^2$ terms will dominate. For example, when $n = 100$:

$$\frac{100^2 - 100}{100^2 + 1} = \frac{99,900}{10,001} = 0.99899 \approx 1 = \frac{100^2}{100^2} $$

So, since the $n^2$ terms dominate, we can simplify by dropping all smaller terms:

$$ L = \lim_{n \rightarrow \infty} \frac{n^2 - n}{n^2 + 1} = \lim_{n \rightarrow \infty} \frac{n^2}{n^2} = \lim_{n \rightarrow \infty} 1 = 1. $$

So, in this casem $a(n)$ and $b(n)$ converge to zero *at the same rate.*


:::{note} Order Notation

Given two sequences, $\{a(n)\}_{n=1}^{\infty}$ and $\{b(n)\}_{n=1}^{\infty}$ that both converge to zero, where:

$$L = \lim_{n \rightarrow \infty} \frac{a(n)}{b(n)} < \infty $$

then $a(n)$ converges to 0 *on the order of* (as fast or faster than) $b(n)$:

$$a(n) = \mathcal{O}(b(n)).$$

The $\mathcal{O}$ sign means *on the order of*.

Equivalently, $a(n) = \mathcal{O}(b(n))$ if and only if there exists some $0 < C < \infty$ and some $m$ such that $a(n) \leq C b(n)$ for all $n \geq m$. That is, if $a(n)$ can be bounded from above by a multiple of $b(n)$ for $n$ sufficiently large.

:::

For example, the sequence $a(n) = n/(1 + n^2)$ goes to zero as fast as $1/n$, so $a(n) = \mathcal{O}(n^{-1})$. 

Anytime you are given a rational function for $a(n)$, you can identify its order of convergence by stripping out everything except the dominating term in the numerator and denominator. These are the terms where $n$ is raised to the largest power. 

For example, the sequence:

$$a(n) = \frac{n^3 - 2 n^2 + 1}{n^6 + 2n} = \mathcal{O}\left( \frac{n^3}{n^6} \right) = \mathcal{O}(n^{-3}).$$

If you can't simplify the limit directly, or by stripping out everything but the dominating terms, then try using L'Hopital's rule.


:::{tip} L'Hopital's Rule

If the functions $f(x)$ and $g(x)$ that are differentiable for all $x$ on an open interval around some $x_*$, and $\lim_{x \rightarrow x_*} f(x) = \lim_{x \rightarrow x_*} g(x) = 0$ or $\pm \infty$, and $\frac{d}{dx} g(x)$ is nonzero on an open interval of $x_*$ excluding $x_*$, then:

$$\lim_{x \rightarrow x_*} \frac{f(x)}{g(x)} = \lim_{x \rightarrow x_*}  \frac{f'(x)}{g'(x)} $$

where $f'(x) = \frac{d}{dx} f(x)$ and $g'(x) = \frac{d}{dx} g(x)$. 

:::

Consider the sequences:

$$a(n) = e^{-n}, \quad b(n) = n^{-m} $$

for some integer $m > 0$. Which converges faster?

$$L = \lim_{n \rightarrow \infty} \frac{a(n)}{b(n)} = \lim_{n \rightarrow \infty} \frac{e^{-n}}{n^{-m}} = \lim_{n \rightarrow \infty} \frac{n^m}{e^n} $$

Let $f(x) = x^m$ and $g(x) = e^x$. Then we can apply L'Hopital's rule to work out the limit. 

1. If $m = 1$:

    $$ L = \lim_{x \rightarrow \infty} \frac{x}{e^x} = \lim_{x \rightarrow \infty} \frac{\frac{d}{dx}x}{\frac{d}{dx}e^x} = \lim_{x \rightarrow \infty} \frac{1}{e^x} = 0. $$

    So, $e^{-n}$ converges faster than $n^{-1}$. That's not surprising. Exponential sequences adecay quickly, while harmonic sequences decay slowly. 

1. If $m = 2$:

    $$ L = \lim_{x \rightarrow \infty} \frac{x^2}{e^x} = \lim_{x \rightarrow \infty} \frac{\frac{d}{dx}x^2}{\frac{d}{dx}e^x} = \lim_{x \rightarrow \infty} \frac{2x}{e^x} = 0 $$

    since we've already shown that $\lim_{x \rightarrow \infty} x/e^x = 0$. 

    So, $e^{-n}$ also converges faster than $n^{-2}$.

1. If $m = 3$:

    $$ L = \lim_{x \rightarrow \infty} \frac{x^3}{e^x} = \lim_{x \rightarrow \infty} \frac{\frac{d}{dx}x^3}{\frac{d}{dx}e^x} = \lim_{x \rightarrow \infty} \frac{3x^2}{e^x} = 0 $$

    since we've already shown that $\lim_{x \rightarrow \infty} x^2/e^x = 0$. 

    So, $e^{-n}$ also converges fatser than $n^{-3}$.

Repeating this argument inductively yields the following conclusion:

:::{important} Exponential is Faster than *All* Powers

The sequence $a(n) = r^n$ for $0 < r < 1$ converges to zero *faster* than the sequence $n^{-\gamma}$ for *any* $\gamma > 0$. 

:::

This rule generalizes our observation about geometric and harmonic sequences. The geometric sequence deacys exponentially, so decays faster than any power law. A distribution with exponential tails will have "lighter" tails than any distribution with power law tails. 

The plot below compares the exponential function $e^{-(n-1)}$ to the power laws we plotted before.

![Power Law to Exponential Comparison.](log_log_exponential.png "Power Law Comparison to Exponential on a Log-Log Plot") 

The exponential clearly converges faster than $n^{-1}$, $n^{-2}$, or $n^{-3}$. It converges faster than $n^{-\gamma}$ for any $\gamma > 0$ since the power law, $n^{-\gamma}$ would be a line with slope $-\gamma$ on the log-log plot, while the exponential is a concave curve whose slope grows steeper as $\log(n)$ increases on the log-log plot. No matter what $\gamma$ you choose, I could pick an $n$ large enough so that the exponential curve is decreasing with faster than $-\gamma$. 

## Ordering Tail Decay Rates

During this week you will practice ordering sequences by their decay rates. We've set up some essential references in this chapter.

Our slowest example is the harmonic sequence, $n^{-1}$, which converges so slowly that its associated series, $\sum_{n = 1}^{\infty} n^{-1}$ diverges.

Our fastest examples are exponential/geometric, $r^n$ for $0 < r < 1$. 

Power laws of the form $n^{-\gamma}$ for $\gamma > 1$ fall in between. They are faster than harmonic, but slower than exponential. They are ordered by their power. 

Whenever you are given a distribution function, it is a good idea to check the tails, and to try and determine the order of convergence of the tail to zero. In essence, try to match its decay rate to the decay rate of some simple reference function (e.g. a power law or an exponential).
