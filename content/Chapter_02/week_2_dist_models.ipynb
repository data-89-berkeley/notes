```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 100 points
np.random.seed(42)
x = np.linspace(0, 5, 100)
# Exponential growth: y = 2 * e^(0.5 * x) + noise
y = 2 * np.exp(0.5 * x) + np.random.normal(0, 0.5, 100)

plt.scatter(x, y, s=10)
plt.xlabel("X (Independent Variable)")
plt.ylabel("Y (Dependent Variable)")
plt.show()
```
