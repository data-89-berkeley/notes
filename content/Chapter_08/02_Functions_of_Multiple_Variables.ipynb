# Integrals of Sums and Products

## Sums

Suppose that $f(x) = a g(x) + b h(x)$ for some constants $a$ and $b$, and some factors $g(x)$ and $h(x)$. What is $\sum_{\text{all } x} f(x)$? What about $\int_{\text{all } x} f(x) dx$? 

Integrals and sums are linear operations. That means that, like the expectation, the integral of a sum is the sum of the integrals, the integral of a linear function is the linear function of the integral, and integrals of linear combinations are linear combinations of integrals. The same rules apply for summations. In all cases, the area under the sum of two curves is the sum of the area under each curve.

:::{note} Linearity of Sums/Integrals

If $f(x) = a g(x) + b h(x)$ then:

$$\begin{aligned} &\sum_{\text{all } x} f(x) = a \sum_{\text{all } x} g(x) + b \sum_{\text{all } x} h(x), \\ &\int_{\text{all } x} f(x) dx = a \int_{\text{all } x} g(x) dx + b \int_{\text{all } x} h(x) dx. \end{aligned}  $$

In other words, the *sum/integral of a linear combination equals the linear combination of the sums/integrals.*

:::

Notice the similarity to linearity of expectation. Expectations are linear since they are weighted averages, and averages are sums/integrals. 

:::{tip} Mixture Example
:class: dropdown

Suppose that $I \sim \text{Bernoulli}(1/3)$ is an indicator random variable. Suppose that, if $I = 0$ then $X \sim \text{Binomial}(100,1/5)$. Suppose that, if $I = 1$, then $X \sim \text{Binomial}(50,4/5)$. What is $\mathbb{E}[X]$?

To find the expectation, first expand the PMF by partitioning:

$$\begin{aligned} \text{PMF}(x) = \text{Pr}(X = x) & = \text{Pr}((X=x,I = 0) \cup (X = x, I = 1)) \\ & = \text{Pr}(X = x,I = 0) + \text{Pr}(X = x,I = 1). \end{aligned}$$

Then, use the multiplication rule to expand each joint probability:

$$ \begin{aligned} \text{PMF}(x) & = \text{Pr}(I =0) \text{Pr}(X = x|I = 0) + \text{Pr}(I = 1) \text{Pr}(X = x|I = 1) \\
& = \frac{1}{3} \text{Pr}(X = x|I = 0) + \frac{2}{3} \text{Pr}(X = x|I = 1). \end{aligned}$$

So, the PMF is a linear combination of the conditional PMF's given $I = 0$ and $I = 1$. This is an example of a *mixture distribution* (see [Section 3.2](#ch3.2)).

It follows that:

$$\mathbb{E}[X] = \sum_{\text{all }x} x \text{PMF}(x) = \sum_{\text{all }x} \frac{1}{3} \left(x \text{Pr}(X = x|I = 0) \right) + \frac{2}{3} \left(x \text{Pr}(X = x|I = 0) \right).  $$

Let $g(x) = x \text{Pr}(X =x|I = 0)$ and $h(x) = x \text{Pr}(X = x|I = 1)$. Then, applying the linearity of sums:

$$\mathbb{E}[X] = \frac{1}{3} \sum_{\text{all }x} x \text{Pr}(X = x|I = 0) + \frac{2}{3} \sum_{\text{all }x} x \text{Pr}(X = x|I = 1). $$

Now, each sum is a weighted average of $x$ against a PMF, so is an expectation. The first is the expected value of a Binomial random variable on 100 trials with success probability $1/5$, so equals $100 \times \frac{1}{5} = 20$. The second is the expected value of a Binomial random variable on 50 trials with success probability $4/5$ so equals $50 \times 4/5 = 40$. Therefore:

$$\mathbb{E}[X] = \frac{20}{3} + \frac{2 \times 40}{3} = \frac{100}{3} = 33.333... $$

Replacing each number with the quantity it represents:

$$\mathbb{E}[X] = \text{Pr}(I = 0) \mathbb{E}[X|I = 0] + \text{Pr}(I = 1) \mathbb{E}[X|I = 1] $$

where the conditional signs inside the expectations indicate an expectation against the conditional distribution specified by the conditioning statement. This is an example of a related property of expectations. We will explore this property more in the future. For now, it's enough to observe that linearity helps break problems into simpler parts. 
:::


## Products

Anytime we go to evaluate an expectation we need to compute a sum, or integral, of a product of two functions (typically, the function of $X$ whose expectation we are interested in, and the distribution function of $X$).

So, suppose $f(x) = g(x) \times h(x)$. What is $\sum_{\text{all } x} f(x)$? What about $\int_{\text{all }x} f(x) dx$?

We'll state the general rule for integrals, provide some examples, then will illustrate an analogous rule for sums. 

### Integration by Parts

:::{note} Integration by Parts

Given $f(x) = g(x) h'(x)$, 

$$\int_{x = a}^b f(x) dx = g(x) h(x) \Big|_{x = a}^b - \int_{a}^b g'(x) h(x) dx$$

where $h'(x) = \frac{d}{dx} h(x)$ and $g'(x) = \frac{d}{dx} g(x)$.
:::

:::{hint} Proof
:class: dropdown

Integration by parts follows simply from the product rule for differentiation. Recall that, if $f(x) = g(x) h(x)$ then:

$$f'(x) = g'(x) h(x) + g(x) h'(x). $$

So, integrating both sides:

$$ \begin{aligned} f(a) - f(b) & = \int_{x = a}^b f'(x) dx = \int_{x = a}^b (g'(x) h(x) + g(x) h'(x)) dx \\ 
& = \int_{x = a}^b g'(x) h(x) dx + \int_{x = a}^b g(x) h'(x) dx.   \end{aligned}$$

So, rearranging:

$$\int_{x = a}^b g(x) h'(x) dx = f(x)|_{x = a}^b - \int_{x = a}^b g'(x) h(x) dx. $$

Finally, expressing $f$ as a product of $g$ and $h$:

$$\int_{x = a}^b g(x) h'(x) dx = g(x) h(x)|_{x = a}^b - \int_{x = a}^b g'(x) h(x) dx. $$

:::

Integration by parts swaps which factor is differentiated. It replaces $h'$ with $h$ inside the integral, and replaces $g$ with $g'$. In essence, we are moving the derivative from $h$ to $g$. To move the derivative we will need to know $h$, which is the *anti*-derivative of $h'$. In probability $h'$ is often a PDF, so $h$ is often a CDF. 

:::{tip} A Mnemonic

This is a long formula. A mnemonic helps.

This rule is often written:

$$\int_{x = a}^b u(x) dv(x) = u(x) v(x) \Big|_{x = a}^b - \int_{a}^b  v(x) du(x) $$

where $dv(x) = \frac{d}{dx} v(x) dx = v'(x) dx$ and $du(x) = \frac{d}{dx} u(x) dx = u'(x) dx$. 

You can remember the right hand side "**u**ltra-**v**iolet **v**oo**doo**" for "$u$ $v$ less $v$ $du$."

:::

To apply integration by parts, always apply the same steps. These are:

1. Write $f$ as a product of two functions: $f(x) = g(x) h'(x)$. Look for a pair of factors where:
    - $g'(x)$ is a simpler function than $g(x)$
    - $h'(x)$ is easy to integrate to recover $h(x)$. If we can't find the anti-derivative, then we won't be able to integrate by parts. 

1. Find $g'(x)  = \frac{d}{dx} g(x)$.

1. Find $h(x) = \int^x h'(s) ds$. 

1. Try to evaluate $\int_{a}^b g'(x) h(x) dx$.

1. Combine terms. 


:::{tip} Example

Suppose that $X \in [0,1]$ and $\text{PDF}(x) \propto g(x) =  x (1 - x)^4$. What is the normalizing constant?

To find the normalizing constant, we should integrate $g(x)$ over all $x$:

$$\int_{x = 0}^1 x (1 - x)^4 dx = \int_{x = 0}^1 u(x) v'(x) dx $$

where $u(x) = x$ and $v'(x) = (1 - x)^4$. Then $du(x) = \frac{d}{dx} u(x) dx = 1 dx = dx$ and $v(x) = \int^x dv(s) = \int^x (1 - x)^4 ds = -\frac{1}{5}(1 - s)^5|^x = -\frac{1}{5}(1 - x)^5$. 

Next, evaluate the integral of $v du$:

$$\begin{aligned} \int_{x = 0}^1 v(x) du(x) & = -\frac{1}{5}  \int_{x = 0}^{1} (1 - x)^5 dx \\
& = \frac{1}{5} \frac{1}{6} (1 - x)^{6}|_{0}^1 \\
& = \frac{1}{30}(0 - 1) = -\frac{1}{30}.  \end{aligned}$$

The "$u v$" term is:

$$u(x) v(x) \Big|_{0}^1 = - \frac{1}{5} x (1 - x)^5 \Big|_{0}^1 = -\frac{1}{5}(0  - 0) = 0. $$

So, combining terms:

$$\int_{x = 0}^1 g(x) = 0 + \frac{1}{30} = \frac{1}{30}. $$

It follows that:

$$\text{PDF}(x) = 30 x (1 - x)^4. $$

To check our work, notice that the value of the integral wouldn't change if we reflected the function $g(x)$ about $x = 0.5$. This swaps $x$ with $1 - x$. Then:

$$\begin{aligned} \int_{x = 0}^1 x^4 (1 - x) dx & = \int_{x = 0}^1 x^4 - x^5 dx \\ 
& = \int_{x = 0}^1 x^4 dx - \int_{x = 0}^1 x^5 dx \\ 
&  = \frac{1}{5} x^5 - \frac{1}{6} x^6 |_{0}^1 \\
& = \frac{1}{5} - \frac{1}{6} = \frac{1}{30}. \end{aligned}$$

:::


:::{tip} Example

Suppose that $X \sim \text{Exponential}(\lambda)$. What is $\mathbb{E}[X]$?

Expanding the expectation as a weighted average gives:

$$\mathbb{E}[X] = \int_{x = 0}^{\infty} x \text{PDF}(x) dx = \int_{x = 0}^{\infty} x \lambda e^{-\lambda x} dx. $$

Set $g(x) = x$ and $h'(x) = \lambda e^{-\lambda x}$. Then $g'(x) = 1$ and $h(x) = - e^{-\lambda x}$. 

Then:

$$\begin{aligned} \int_{x = 0}^{\infty} g'(x) h(x) dx & = - \int_{x = 0}^{\infty} e^{-\lambda x} dx
= \frac{1}{\lambda} e^{-\lambda x}|_{0}^\infty \\
& =  \frac{1}{\lambda}(0 - 1) = -\frac{1}{\lambda}. \end{aligned} $$

The "$g h$" term is:

$$-x e^{-\lambda x} \Big|_0^{\infty} = 0 - 0 = 0. $$

Therefore:

$$\int_{x = 0}^{\infty} x \lambda e^{-\lambda x} dx = 0 - \left(-\frac{1}{\lambda}\right) = \frac{1}{\lambda}. $$

So, if $X \sim \text{Exponential}(\lambda)$ then $\mathbb{E}[X] = \lambda$. 


:::

#### Tail Integrals

Any time we ask for the expectation of a continuous, nonnegative random variable we need to evaluate the integral $\int_{x = 0}^{b} x \text{PDF}(x) dx$ where $b$ is the upper endpoint of the support. In this case integration by parts produces an alternate formula for the expectation using the area under the survival function.

Setting $g(x) = x$ gives $g'(x) = 1$. Setting $h'(x) = \text{PDF}(x)$ sets $h(x) = \text{CDF}(x)$. Then, integrating by parts:

$$\begin{aligned} \int_{x = 0}^b x \text{PDF}(x) dx & = x \text{CDF}(x) \Big|_{0}^b - \int_{x = 0}^b \text{CDF}(x) dx \\
& =  [b \times \text{CDF}(b) - 0 \times \text{CDF}(0)] - \int_{x = 0}^b \text{CDF}(x) dx  \\
& = [b \times 1 - 0 \times 0] - \int_{x = 0}^b \text{CDF}(x) dx  \\
& = b - \int_{x = 0}^b \text{CDF}(x) dx \\
& = \int_{x = 0}^b 1 - \text{CDF}(x) dx \\
& = \int_{x = 0}^b 1 - \text{Pr}(X \leq x) dx \\
& = \int_{x = 0}^b \text{Pr}(X > x) dx. \end{aligned}$$

This formula holds for any $b$, so holds for any nonnegative random variable. We can extend the upper bound of the integral to infinity since $\text{Pr}(X > x) = 0$ if $x$ is above the upper bound of the support.

:::{note} The Tail Integral Formula

If $X \geq 0$ then:

$$\mathbb{E}[X] = \int_{x = 0}^\infty \text{Pr}(X > x) dx. $$

In other words, *the expectation of a nonnegative random variable is the area under its survival function.*

:::

For example, if $X \sim \text{Exponential}(\lambda)$ then $\text{Pr}(X > x) = e^{-\lambda x}$ so $\mathbb{E}[X] = \int_{x = 0}^{\infty} e^{-\lambda x} = \frac{1}{\lambda}.$ 

### Summation by Parts

Most integral rules match a rule for sums. The integration by parts strategy can also be used for sums of products.

Consider a sum of the form

$$\sum_{x = 1}^n g(x) h(x) $$

for $n$ possibly infinite. 

The discrete analog to a derivative is a difference. Let:

$$g'(x) = g(x) - g(x - 1), \quad g(0) = 0. $$

Then, just as derivatives are inverted by integrals, differences are inverted by running sums:

$$g(x) = (g(x) - g(x - 1)) + (g(x - 1) - g(x - 2)) + ... (g(2) - g(1)) + g(1) = \sum_{y=1}^x g'(y). $$

So, we can write our original sum:

$$\sum_{x = 1}^n g(x) h(x) = \sum_{x = 1}^n \left[\sum_{y = 1}^x g'(y) \right] h(x) = \sum_{x = 1}^n \sum_{y=1}^x g'(y) h(x). $$

To recover the discrete analog to integration by parts, reverse the order of the sum. First, notice that the sum runs over all pairs $x, y$ such that $1 \leq y \leq x \leq n$. We can write this sum:

$$\sum_{x = 1}^n \sum_{y=1}^x g'(y) h(x) = \sum_{1 \leq y \leq x \leq n} g'(y) h(x) = \sum_{y = 1}^n \sum_{x = y}^n g'(y) h(x).$$

Since $g'(y)$ does not depend on $x$:

$$\sum_{y = 1}^n \sum_{x = y}^n g'(y) h(x) = \sum_{y = 1}^n g'(y) \left[\sum_{x = y}^n h(x) \right].$$

The sum, $\sum_{x = y}^n h(x)$ is analogous to an anti-derivative. To make the analogy explicit, let $H(y) = \sum_{x = 1}^{y-1} h(x)$. Then $\sum_{x = y}^n h(x) = H(n+1) - H(y)$ so:

$$\begin{aligned} \sum_{y = 1}^n g'(y) \left[\sum_{x = y}^n h(x) \right] & = \sum_{y = 1}^n g'(x) [H(n+1) - H(y)] \\ & = H(n+1) \left( \sum_{y = 1}^n g'(x) \right) - \sum_{y=1}^n g'(y) H(y). \end{aligned}$$

Finally, since $\sum_{y=1}^n g'(x) = g(n) - g(0)$:

$$\sum_{x = 1}^n g(x) h(x) = g(n) H(n+1) - \sum_{y=1}^n g'(y) H(y). $$

:::{note} Summation by Parts

Given two sequences $\{g(x)\}_{x = 1}^n$ and $\{h(x)\}_{x = 1}^n$ where, by convention $g(0) = 0$. Let $g'(x) = g(x) - g(x - 1)$ and $H(y) = \sum_{x < y} h(x)$. Then:

$$\sum_{x = 1}^n g(x) h(x) = g(n) H(n+1) - \sum_{y=1}^n g'(y) H(y) $$

:::

The summation by parts formula is perfectly analogous to the integration by parts formula. You'll prove integration by parts on your homework using the same steps, but replacing differencing with derivatives, and summation with integration.



#### Tail Sums

Summation by parts is most often used to find the expectation of count valued random variables.

Suppose that $X \in \{0,1,2,3,...,n\}$. Then:

$$\mathbb{E}[X] = \sum_{x = 0}^{n} x \text{PMF}(x) $$

where $n$ is possibly infinite. 

Then, let:

$$g(x) = x, \quad h(x) = \text{PMF}(x).$$

Then:

$$g'(x) = g(x) - g(x - 1) = x - (x - 1) = 1$$

and:

$$H(y) = \sum_{x < y} h(x) = \sum_{x < y} \text{PMF}(x) = \text{CDF}(y-1).$$

It follows that:

$$\begin{aligned} \mathbb{E}[X] & = n \times \text{CDF}(n)- \sum_{y=1}^n 1 \times \text{CDF}(y - 1) \\
& = n \times 1 - \sum_{y=1}^n 1 \times \text{CDF}(y-1) = n - \sum_{y=1}^n \text{CDF}(y-1) \\
& = \sum_{y = 1}^n 1 - \text{CDF}(y - 1) = \sum_{y = 1}^{n} (1 - \text{Pr}(X < y)) \\
& = \sum_{y = 1}^{n} \text{Pr}(X \geq y). \end{aligned}$$

:::{note} The Tail Sum Formula

If $X \in \{0,1,2,3,...,n\}$ then:

$$\mathbb{E}[X] = \sum_{y = 1}^{n} \text{Pr}(X \geq y) $$

In other words, *the expectation of a count valued random variable is the area under its survival function.*

:::

The tail sum formula is often proved using a different algebraic argument. Consider the sum:

$$\mathbb{E}[X] = \sum_{x = 0}^{n} x \text{PMF}(x) = \sum_{x = 1}^{n} x \text{PMF}(x). $$

For concision, let $p(x) = \text{PMF}(x)$. Then the sum has the form:

$$\mathbb{E}[X] = 1 \times p(1) + 2 \times p(2) + 3 \times p(3) + ... n \times p(n). $$

We can organize this sum:

$$\begin{aligned} \mathbb{E}[X] = & p(1) + ... \\
& p(2) + p(2) + ... \\
& p(3) + p(3) + p(3) + ... \\
& \vdots \\
& p(n) + p(n) + p(n) + p(n) +... p(n)\end{aligned}$$

Summing down the columns instead of across the rows returns the tail sum formula:

$$\begin{aligned} \mathbb{E}[X] & = \left(\sum_{x = 1}^n p(x) \right) + \left(\sum_{x = 2}^n p(x) \right) + \left(\sum_{x = 3}^n p(x) \right) + ... + (p(n)) \\
& = \text{Pr}(X \geq 1) + \text{Pr}(X \geq 2) + ... + \text{Pr}(X \geq n) = \sum_{y=1}^n \text{Pr}(X \geq y). \end{aligned}$$

This proof is less general, but more transparent.


:::{tip} Example

Suppose that $X \sim \text{Geometric}(p)$. What is $\mathbb{E}[X]$?

If $X$ is a geometric random variable then $X$ is count valued. So, using the tail sum formula:

$$\mathbb{E}[X] = \sum_{x = 1}^{\infty} \text{Pr}(X \geq x).$$

The chance that a geometric random variable is greater than *or equal* to a threshold $x$, is the chance that, $x - 1$ independent, identical, binary trials all fail. The trials have success probability $p$, so the chance they each fail is $1 - p$. The chance of $x - 1$ failures is $(1 - p)^{x - 1}$. Therefore:

$$\mathbb{E}[X] = \sum_{x = 1}^{\infty} (1 - p)^{x - 1} = \sum_{n = 0}^{\infty} q^n$$

where $q = 1 - p$. 

Since the sum is a geometric series (see [Section 5.1](#ch5.1)) it converges to:

$$\mathbb{E}[X] = \sum_{n = 0}^{\infty} q^n = \frac{1}{1 - q} = \frac{1}{p}.$$

So, the expectation of a geometric random variable is the reciprocal of its success probability. 

This result makes intuitive sense. If the chance of success is 1/100, then it is sensible that the expected number of trials until a success is 100. 

:::
