# Probability Fundamentals

## Chapter Overview

This chapter is essential. It is the most important in the book. It introduces the fundamental rules of chance that we will study through the rest of this course. These establish the mathematics that define probability as a subject. 

The chapter will:
1. Describe **outcomes** and **events** as **sets**

1. Illustrate how to convert any logical statement regarding events into operations on sets

1. Introduce the simplest chance model: chances for **equally likely outcomes.** 
    - In this setting, the **probability** of an event equals the **proportion** of all outcomes consistent with the event. 
    - For example, the probability that a fair six-sided die lands on a prime number is 4/6 since there are 6 possible rolls, 4 of which are prime (rolls 1,2,3,5).

1. We will then generalize rules observed for equally likely events, to compute the probability of any event that can be defined as a logical statement regarding outcomes. 
    - We will highlight **rules for *not*, *or*, *if*, and *and* statements**.
    
1. We will then use these rules to: 
    - **compute and bound chances**, and
    - show how to update chances when we **restrict the possible set of outcomes by conditioning**.
