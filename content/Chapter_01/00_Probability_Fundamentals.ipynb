# Probability Fundamentals

## Chapter Overview

This chapter is essential. It is the most important in the book. It introduces the fundamental rules of chance that we will study for the next 14 weeks.

The chapter will:
1. Describe **outcomes** and **events** as **sets** ([Section 1.1](ch1.1))

- Illustrate how to convert any logical statement regarding events into operations on sets ([Section 1.1](ch1.1))

1. Introduce the simplest chance model: chances for **equally likely outcomes.** ([Section 1.2](ch1.2))
    - In this setting, the **probability** of an event equals the **proportion** of all outcomes consistent with the event. 
    - For example, the probability that a fair six-sided die lands on a prime number is 4/6 since there are 6 possible rolls, 4 of which are prime (rolls 1,2,3,5).

1. Generalize rules observed for equally likely events to compute the probability of any event that can be defined as a logical statement regarding outcomes. ([Section 1.3](ch1.3))
    - We will highlight  **rules for**
        - ***not*, *or*,** ([Section 1.3](ch1.3))
        - ***if*,** and ([Section 1.5](ch1.5))
        - ***and* statements** ([Section 1.4](ch1.4)).
    
1. Use these rules to: 
    - **compute and bound chances** ([Section 1.3](ch1.3) and [Section 1.4](ch1.4)), and
    - show how to update chances when we **restrict the possible set of outcomes by conditioning** ([Section 1.5](ch1.5)). 
    - Define **independent** events ([Section 1.6](ch1.6)).
