# Combinatorics Helper

Many problems in discrete probability involve counting arrangements. This is especially true for models where outcomes are equally likely, since, when outcomes are equally likely, probabilities are ratios of counts.

We will not emphasize counting problems heavily in this class, so will try to avoid heavy combinatorics. However, there are two counting problems that are so common, we might as well learn the combinatorics. Both involve counting the number of ways we can arrange a list of items. They differ by whether or not the items are distinguishable from each other.

## Distinguishable Items

Suppose that you have a list of $n$ items. For specificity, imagine a list of $n$ cards. Suppose that the cards are all distinguishable from each other. For example, all the cards have a unique number.

:::{note} Permutation Definition
A **permutation** is a distinct arrangement of a list of $n$ distinguishable.
:::

üõ†Ô∏è How many ways are there to arrange the cards?

:::{hint} Solution
:class: dropdown

We solved this problem back in [Section 1.2](ch1.2) when we wanted to find the chance of drawing two aces in a row from a well-shuffled deck.

Remember, when counting, it is often best to think in sequence.

There are $n$ options for the first draw. After the first draw, there are $n - 1$ options. After the first two draws there are $n - 2$ options for the third. So, repeating the pattern, there are:

$$n \times (n-1) \times (n-2) \times ... 3 \times 2 \times 1 = n!$$

So, for any list of $n$ distinguishable objects, there are always $n!$ distinct permutations.

:::

:::{note} Counting Permuations
There are always $n!$ distinct permutations of $n$ objects.
:::

## Two Types of Items

What if not all the items can be distinguished?

Then $n!$ overcounts, because it counts rearrangements that cannot be distinguished.

For example, suppose that I have five cards. The front side of three of the cards are red, $R$, and the front of the remaining two are blue, $B$. The back side of the cards are numbered from 1 to 5 so that the red cards are labelled 1, 2, and 3, and the blue cards are labelled 4 and 5.

So, when the cards are front side up, I can't tell the order of the red cards apart. Nor can I tell the order of the blue cards apart. 

In this case, we can list all the possible distinct arrangements:

$$\{RRRBB,RRBRB,RRBBR,RBRRB,RBRBR,RBBRR,BRRRB,BRRBR,BRBRR,BBRRR\} $$

There are 10 possible arrangements. Where did the number 10 come from?

:::{hint} Solution
:class: dropdown
Here's a nice argument. Think about each arrangement listed above. For example, think about the arrangement $RRRBB$. This sequence could correspond to any permutation of the cards so that the first three are red. We can spell these out by peaking at the back side of the cards:

Front | R | R  | R  | B | B |
:----:|:---------|:---------|:---------|:---------|:---------|
Back | 1 | 2 | 3 | 4 | 5 |
Back | 1 | 2 | 3 | 5 | 4 |
Back | 1 | 3 | 2 | 4 | 5 |
Back | 1 | 3 | 2 | 5 | 4 |
Back | 2 | 1 | 3 | 4 | 5 |
Back | 2 | 1 | 3 | 5 | 4 |
Back | 2 | 3 | 1 | 4 | 5 |
Back | 2 | 3 | 1 | 5 | 4 |
Back | 3 | 1 | 2 | 4 | 5 |
Back | 3 | 1 | 2 | 5 | 4 |
Back | 3 | 2 | 1 | 4 | 5 |
Back | 3 | 2 | 1 | 5 | 4 |

Take a careful look at the structure of this table. Notice that, for every order of the red cards there are $2 = 2!$ ways to order the blue cards. For every order of the blue cards, there are $6 = 3!$ ways to order the red cards. 

So, if we fix the order of red and blue cards, then there are $3! \times 2!$ ways to arrange the cards.

Imagine, now, that instead of seeing the front of the cards, you see the back of the cards. Then, you can distinguish all 5, so would be able to distinguish $5!$ distinct sequences. But, those $5!$ sequences can be broken into sets of size $3! \times 2!$, where all permutations within each set order the colors (front side of the cards) identically. 

Therefore, the number of distinguishable arrangements is:

$$\frac{5!}{3! \times 2!} = \frac{(\text{total number of cards})!}{(\text{number of red cards})!(\text{number of blue cards})!} $$

:::

The solution given above is an example of the next combinatorial operator after a factorial.

:::{note} The Binomial Coefficient

The **binomial coefficient** $\{n \choose m}$ is the ratio:

$$\{n \choose m} = \frac{n!}{m!(n-m)!}$$

:::

We usually read a binomial coefficient as $n$ *choose* $m$. We use the word *choose* since the binomial coefficient counts the number of ways we could choose $m$ items from a collection of $n$, if we don't care the order we choose in. 

This is one common application of the binomial coefficient. Suppose I have a list of $n$ names, and want to form a poll group by randomly selecting $m$ individuals. If I don't care about the order in which I pick, then every individual is simply assigned to the "picked" class, or the "not picked" class. Choosing $m$ individuals from all $n$ is then the same as assigning each individual a $P$ or a $N$. If we list the individuals in order, then each selection is a string of length $n$, containing $m$ letters $P$, and $(n - m)$ letters $N$. 

For example, the string $NNPNP$ means we formed out poll group by choosing individuals 3 and 5. Then, the number of possible groups is the number of $n$ letter words made up of $m$ P's and $(n-m)$ N's. This is, the number of ways to rearrange a list of objects that are divided into two classes of fixed size, so is the same as the problem we solved above.

:::{note} Choose

The number of distinct ways to:

1. Order $n$ objects divided into two classes of sizes $m$ and $n-m$
1. Choose $m$ objects from a list of $n$ distinct objects, disregarding the order of the draw

is always:

$$\{n \choose m} = \frac{n!}{m!(n-m)!}$$

:::
